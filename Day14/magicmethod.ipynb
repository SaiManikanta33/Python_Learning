# without   __str__()

```python
class Student:
    def __init__(self,name):
        self.name = name
student = Student("Sai")
print(student)
```

# with   __str__()

```python
class student:
    def __init__(self,name):
        self.name = name
    def __str__(self):
        return f"Student(Name={self.name})"
Student = student("Sai")
print(Student)
```

# Customize len()

```python
class team:
    def __init__(self,members):
        self.members = members
    def __len__(self):
        return len(self.members)
Team = team(["Sai","Mani"])
print(len(Team))
```

#   operator overloading with __add__()

```python
class wallet:
    def __init__(self,amount):
        self.amount = amount
    def __add__(self,other):
        return wallet(self.amount + other.amount)
    def __str__(self):
        return f"₹{self.amount}"
Wallet1 = wallet(500)
Wallet2 = wallet(800)
Wallet3 = wallet(1000)

Wallet4 = Wallet1 + Wallet2 + Wallet3
print(Wallet4)
```
