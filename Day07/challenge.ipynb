Personal Notes Manager

Features:

Add a note.
-View all notes.
-Save notes to notes.txt.
-Handle missing file errors gracefully.

Example menu:

===== Notes Manager =====

1. Add Note
2. View Notes
3. Exit

Enter choice:

Use:

-Functions
-File handling
-Exception handling
-with open()

```python
#file = open("Notes.txt","x")

def notes_handle():
    with open("Notes.txt", "w") as file:
        file.write("1.Add note\n")
        file.write("2.View notes\n")
        file.write("3.Exit")
    
    choice = input("\nEnter choice: ")
    
    if choice == '1':
        notes = input("Add your note.\n")
        with open("Notes.txt","w") as file:
            file.write(notes + "\n")
    elif choice=='2':
        with open("Notes.txt","r") as file:
            print(file.read())
    elif choice=='3':
        print("Exiting...")
    else:
        print("Invalid Input")
    try:
        with open("Notes.txt","r") as file:
            print(file.read())
    except FileNotFoundError:
        print("File not found.")
    finally:
        print("Your Notes is Ready.")
notes_handle()
```
