Challenge

Extend the project to:

-Search for a device by hostname.
-Remove a device.
-Prevent duplicate IP addresses.
-Display devices sorted by hostname.

```python
class Device:
    def __init__(self,hostname,ip):
        self.hostname = hostname
        self.ip = ip
        
    def __str__(self):
        return f"{self.hostname} ({self.ip})"
    
class AssetInventory:
    def __init__(self):
        self.devices = []
        
    def add_device(self,device):
        for existing_device in self.devices:
            if existing_device.ip == device.ip:
                print(f"Cannot Add {device.hostname} and {device.ip}")
                return False
        self.devices.append(device)
        return True
    
    def __len__(self):
        return len(self.devices)
    
    def display(self):
        for device in self.devices:
            print(device)
            
    def search_device(self,hostname):
        for device in self.devices:
            if device.hostname.lower() == hostname.lower():
                return device
            return None
        
    def remove_device(self, hostname):
        device_to_remove = self.search_device(hostname)
        if device_to_remove:
            self.devices.remove(device_to_remove)
            print(f"🗑️ Removed device: {hostname}")
            return True
        print(f"⚠️ Device not found: {hostname}")
        return False
    
    def display_sorted(self):
        sorted_devices = sorted(self.devices, key=lambda x: x.hostname.lower())
        for device in  sorted_devices:
            print(device)
            
inventory = AssetInventory()
print("--------Adding Devices --------")
inventory.add_device(Device("Firewall","192.168.1.1"))
inventory.add_device(Device("WebServer","192.168.1.100"))
inventory.add_device(Device("DatabaseServer", "192.168.1.50"))
print("\n---------- Prevent duplicate Ip addresses")
inventory.add_device(Device("Webserver", "192.168.1.50"))

print("\n--- 3. Testing Case-Insensitive Search ---")
found = inventory.search_device("webserver")
if found:
    print(f"🔍 Found: {found}")
else:
    print("🔍 Device not found.")

print("\n--- 4. Testing Alphabetical Sorting ---")
inventory.display_sorted()

print("\n--- 5. Testing Removal ---")
inventory.remove_device("WebServer")

print("\n--- 6. Final Inventory State ---")
inventory.display()
print("Total Devices:", len(inventory))
```

Security Event Counter

```python
class SecurityEvent:
    def __init__(self,event_type):
        self.event_type = event_type
        
    def __str__(self):
        return f"Event:{self.event_type}"
class EventLog:
    def __init__(self):
        self.events = []
        
    def add_event(self,event):
        
        self.events.append(event)
    def __len__(self):
        return len(self.events)
    
    def count(self):
        counts = {}
        for event in self.events:
            if event.event_type in counts:
                counts[event.event_type] +=1
            else:
                counts[event.event_type] = 1
        return counts
    
    def search_event(self, keyword):
        results = []
        for event in self.events:
            if keyword.lower() in event.event_type.lower():
                results.append(event)
        return results
    
    def export_to_text(self,filename="Security_logs.txt"):
        try:
            with open(filename,"w") as file:
                file.write("=== Security Log Export ===\n")
                for event in self.events:
                    file.write(f"{event}\n")
            print(f"Successfully exported {len(self.events)} events to {filename}")
        except IOError as e:
            print(f"Erroe writing to file:{e}")
    
log = EventLog()

log.add_event(SecurityEvent("Failed Login"))
log.add_event(SecurityEvent("Port Scan"))
log.add_event(SecurityEvent("Malware Detected"))
log.add_event(SecurityEvent("Port Scan"))

print("Event Logs:",len(log))

for event in log.events:
    print(event)
    
counts = log.count()
for event_type, total in counts.items():
    print(f"{event_type}: {total}")
    
search_results = log.search_event("Login")

if search_results:
    for match in search_results:
        print(f"🎯 Found: {match}")
else:
    print("❌ No matching events found.")
    
log.export_to_text("My_events.txt")
```
