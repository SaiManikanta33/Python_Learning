```python
class Book(object):
    def __init__(self, title):
        self.title = title
    def info(self):
        print("Book:", self.title)

class EBook(Book):
    def __init__(self, title, size):
        super(EBook, self).__init__(title)
        self.size = size
    def info(self):
        print("Book:", self.title)
        print("Size:", self.size, "MB")

class printedbook(EBook):
    def __init__(self, title, size, pages):
        super(printedbook, self).__init__(title, size)
        self.pages = pages
    def info(self):
        print("Book:", self.title)
        print("Pages:", self.pages)

class audiobook(printedbook):
    def __init__(self, title, pages, size, audio):
        super(audiobook, self).__init__(title, size, pages)
        self.audio = audio
    def info(self):
        print("Book:", self.title)
        print("Size:", self.size, "MB")
        print("Pages:", self.pages)
        print("Audio:", self.audio)

ebook = audiobook("Python OOPS", 300, 20, "MP3")
ebook.info()
```
