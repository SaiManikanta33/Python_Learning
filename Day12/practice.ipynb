Exercise 2

Create a Person class.

Attributes:

name
age

Create a Student class that adds:

branch

Print all details.

```python
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
class student(person):
    def __init__(self,name,age,branch):
        super().__init__(name,age)
        self.branch = branch
Student = student("sai",19,"Cyber")
print(Student.name)
print(Student.age)
print(Student.branch)
```

Create an Employee class.

Override the work() method in the Manager class.

Example:

Employee is working

↓

Manager is managing the team

```python
class employee:
    def work(self):
        print("Employee is working")
class manager(employee):
    def work(self):
        print("Manager is managing the team")
obj = manager()
obj.work()
```

Exercise 5

Create a multilevel inheritance example.

Person

↓

Employee

↓

Manager

Display details from all levels.

```python
class person:
    def display(self):
        print("A person")
class employee(person):
    def Display(self):
        super().display()
        print("A employee")
class manager(employee):
    def disp(self):
        super().Display()
        print("A Manager")
obj = manager()
obj.disp()
        
```
