Requirements:

Private balance
deposit()
withdraw()
get_balance()

Prevent withdrawals greater than the balance.

```python
class BankAccount:
    def __init__(self,balance):
        self.__balance = balance
    def deposit(self,amount):
        self.__balance += amount
        return self.__balance
    def withdraw(self,amount):
        if self.__balance > amount:
            self.__balance -= amount
        return amount
    def get_balance(self):
        return self.__balance
acc = BankAccount(5000)
print(acc.get_balance())
depositt = int(input("Enter your deposit money:"))
acc.deposit(depositt)
print(acc.get_balance())
acc.withdraw(500)
print(acc.get_balance())
```

```python
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = float(balance)

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive")
        self.__balance += amount
        return self.__balance

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive")
        if amount > self.__balance:
            raise ValueError("Insufficient funds")
        self.__balance -= amount
        return amount

    def get_balance(self):
        return self.__balance

# replace the existing acc with a corrected account and demonstrate usage
acc = BankAccount(5000)
print(acc.get_balance())   # 5000.0
acc.deposit(1000)
print(acc.get_balance())   # 6000.0
try:
    acc.withdraw(7000)
except ValueError as e:
    print("Withdrawal failed:", e)
print(acc.get_balance())   # remains 6000.0
```

Exercise 3

Create an abstract class:

Shape

Abstract method:

area()

Implement:

Rectangle
Circle

Calculate and print their areas.

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width = width
        self.height = height
    def area(self):
        area = self.width * self.height
        print(f"Reactangle of area:{area}")
        return area
arr = Rectangle(10,20)
arr.area()
```

Exercise 4

Create:

Payment

↓

UPI
CreditCard
NetBanking

Each should implement:

pay()

Print a different message for each payment type.

```python
from abc import ABC,abstractmethod
class payment:
    @abstractmethod
    def pay(self,amount):
        pass
class UPI(payment):
    def pay(self,amount):
        print(f"The transaction {amount} sucess through UPI")
class creditcard(payment):
    def pay(self,amount):
        print(f"The transaction {amount} success via credit card")
class netbanking(payment):
    def pay(self,amount):
        print(f"Redirecting to netbanking {amount}")
tran =UPI()
tran.pay(500)
tran2 = creditcard()
tran2.pay(1000)
tran3 = netbanking()
tran3.pay(5000)
```

Create a class:

Student

Private attributes:

name
marks

Provide methods to:

Set marks
Get marks
Determine if the student passed (marks ≥ 35)

```python
class Student:
    def __init__(self,name,marks):
        self.__name = name
        self.__marks = marks
    def get_marks(self):
        return self.__marks
    def set_marks(self,new_marks):
        if 0 <= new_marks <= 100:
            self.__marks = new_marks
    def get_marks(self):
        return self.__marks
    def has_passed(self):
        return self.__marks >= 35
res = Student("Sai",90)
print(f"Marks: {res.get_marks()}")    
print(f"Passed: {res.has_passed()}")
res.set_marks(95)
print(res.get_marks())
```
