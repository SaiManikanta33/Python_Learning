```python
class networkdevice:
    def __init__(self,ip):
        self.ip = ip
    def connect(self):
        print(f"Connecting to {self.ip}")
class firewall(networkdevice):
    def block(self):
        print("Blocking suspicious traffic")
fw = firewall("192.168.1.1")
fw.connect()
fw.block()
```

```python
class networkdevice:
    def device_type(self):
        print("Devices")
class Firewall(networkdevice):
    def device_type(self):
        print("Firewall Device")
class Router(Firewall):
    def device_type(self):
        super().device_type()
        print("Router Device")
class Switch(Router):
    def device_type(self):
        super().device_type()
        print("Switch Device")
obj = Switch()
obj.device_type()
        
```
