Add a method to determine if the student is eligible for an internship (e.g., CGPA ≥ 7.5)

```python

class student():
    def __init__(self, Rno, college, cgpa):
        self.Rno = Rno
        self.college = college
        self.cgpa = cgpa
    def display(self):
        print("------Student Details ------")
        print("Roll Number : ", self.Rno)
        print("College : ", self.college)
        print("CGPA : ", self.cgpa)
    def eligibility(self):
        if self.cgpa >= 7.5:
            print(f"Student eligible for InternShip with {self.cgpa} CGPA")
        else:
            print("Not eligible for internShip")

Student = student("25985A4608", "Raghu Engg College", 8.18)
Student.display()
Student.eligibility()
```
