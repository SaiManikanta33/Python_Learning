Exercise 3

Create two sets:

set1 = {22, 80, 443}
set2 = {21, 22, 25, 80}

Print:

-Union
-Intersection
-Difference

```python
set1 = {22, 80, 443}
set2 = {21, 22, 25, 80}
print(set1 | set2)
print(set1 & set2)
print(set1 - set2)
```

Exercise 5

Store marks in a dictionary.

marks = {
    "Python": 90,
    "Networking": 85,
    "Linux": 88
}

print:

Subject names
Marks
Average mark

```python
marks = {
    "Python": 90,
    "Networking": 85,
    "Linux": 88
}

for subject, score in marks.items():
    print("Subject:", subject, "Mark:", score)

average = sum(marks.values()) / len(marks)
print("Average:", average)
```
