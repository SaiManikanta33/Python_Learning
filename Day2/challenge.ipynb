🧩 Mini Challenge

Create a simple Student Information Formatter.

Input:

Name
Age
Branch
College

Output:

========== Student Profile ==========
Name    : Sai Manikanta
Age     : 19
Branch  : Cyber Security
College : Raghu Engineering College
=====================================

Use:

input()
f-strings
len()
.title()

```python
name=input("name:").title()
age=int(input("age:"))
branch=input("branch:")
college=input("college:")
print("========== Student details ===========")
print(f"Name    : {name}")
print(f"Age     : {age}")
print(f"Branch  : {branch}")
print(f"College : {college}")
print("=====================================")
```
