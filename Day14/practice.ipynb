Exercise 1

Create a Book class.

Implement:

__str__()

Print:

Book: Python Basics

```python
class Book:
    def __init__(self,bookname):
        self.bookname = bookname
    def __str__(self):
        return f"Book: {self.bookname}"
book = Book("Python Basics")
print(book)
```

Exercise 2

Create a Playlist class.

Implement:

__len__()

Return the total number of songs.

```python
class playlist:
    def __init__(self,songslist):
        self.songslist = songslist
    def __len__(self):
        return len(self.songslist)
play = playlist(["Pushpa","RRR","SpyderMan"])
print(len(play))
```

Exercise 3

Create a Money class.

Overload the + operator.

Example:

m1 = Money(100)
m2 = Money(250)

print(m1 + m2)

Output:

350

```python
class money:
    def __init__(self,amount):
        self.amount = amount
    def __add__(self,other):
        return money(self.amount + other.amount)
    def __str__(self):
        return f"{self.amount}"
m1 = money(100)
m2 = money(250)
print(m1+m2)
```

Exercise 4

Create:

Computer

↓

CPU
RAM

Use composition.

Display:

CPU Ready
RAM Loaded
Computer Started

```python
class CPU:
    def __str__(self):
        return "CPU Ready"
class RAM:
    def __str__(self):
        return "Ram Loaded"
class Computer:
    def __init__(self):
        self.cpu = CPU()
        self.ram = RAM()
    def __str__(self):
        return f" {self.cpu} \n {self.ram} \n Compuer Started"
comp = Computer()
print(comp)
```

Exercise 5

Create:

Department

↓

Professor

Use aggregation.

Print the professor's details through the department.

```python
class Professor:
    def __init__(self,name,id):
        self.name = name
        self.id = id
class Department:
    def __init__(self,professor):
        self.professor = professor
professor = Professor("Sai",101)
dept = Department(prof)
print(dept.professor.name)
print(dept.professor.id)
```
