Password Generator

Challenge:
-Let the user choose the password length.
-Ensure the password includes at least one uppercase letter, one lowercase letter, one digit, and one special character.

```python
import random
import string
length=int(input("Enter password length:"))
if length < 4:
    print("Password length must be at least 4.")
else:
    password=[random.choice(string.ascii_uppercase),
              random.choice(string.ascii_lowercase),
              random.choice(string.digits),
              random.choice(string.punctuation)
              ]
    all_chars=(string.ascii_uppercase+string.ascii_lowercase+string.digits+string.punctuation)
    for _ in range(length - 4):
        password.append(random.choice(all_chars))
    random.shuffle(password)
    password="".join(password)
    print("Generated password:",password)
```
