Exercise 3

Create a generator that yields only even numbers between 1 and 20.

```python
def even_numbers():
    for n in range(2, 21, 2):
        yield n

for num in even_numbers():
    print(num)
```

Exercise 4

Create a generator expression that generates cubes from 1 to 10.

```python
cubes = (n ** 3 for n in range(1, 11))
for num in cubes:
    print(num)
```

Exercise 5

Use sum() on a generator expression to calculate the sum of squares from 1 to 10.

```python
result = sum(x * x for x in range(1, 11))

print(result)
```
