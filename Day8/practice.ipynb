Use a lambda function to find the cube of a number.

```python
cube = lambda x:x ** 3
print(cube(5))
```

Use map() to convert a list of temperatures from Celsius to Fahrenheit.

```python
celcius = [0,20,30,40]
fahrenheit = list(map(lambda c:(c * 9/5) + 32,celcius))
print(fahrenheit)
```

Use filter() to find all numbers greater than 50.

```python
num=[12,56,78,34,99,45]
result = list(filter(lambda x: x > 50,num))
print(result)
```

Use reduce() to multiply all numbers in a list

```python
from functools import reduce
num=[1,2,3,4]
product = reduce(lambda a,b:a*b,num)
print(product)
```

Create a list comprehension that generates the squares of numbers from 1 to 20 that are divisible by 3.

```python
n= [x for x in range(1,20) if x %3==0]
print(n)
```
