Student Grade Calculator

Requirements:

Create the following functions:

-get_marks()
-calculate_average()
-find_grade()
-display_result()

Example:

Python: 90
Networking: 85
Linux: 95

Average: 90

Grade: A

Grade Rules:

-90–100 → A
-75–89 → B
-60–74 → C
-Below 60 → Fail

```python
def get_marks():
    sub1 = float(input("Sub1:"))
    sub2 = float(input("Sub2:"))
    sub3 = float(input("Sub3:"))
    return sub1, sub2, sub3


def calculate_average(m1, m2, m3):
    total = m1 + m2 + m3
    average = total / 3
    return average


def find_grade(average):
    if average >= 90:
        return 'A'
    elif average >= 75:
        return 'B'
    elif average >= 60:
        return 'C'
    else:
        return 'Fail'


def display_result(average, grade):
    print(f"Average: {average:.2f}")
    print(f"Grade: {grade}")


if __name__ == "__main__":
    mark1, mark2, mark3 = get_marks()
    print(f"Python: {mark1}")
    print(f"Networking: {mark2}")
    print(f"Linux: {mark3}")
    ave = calculate_average(mark1, mark2, mark3)
    gra = find_grade(ave)
    display_result(ave, gra)
```
