Log File Reader

Create a file named logs.txt

Read it line by line

Challenge

Print only log entries containing "ERROR".

```python
def read_logs(filename):
    with open(filename,"r") as file:
        for line in file:
            yield line.strip()
for log in read_logs("logs.txt"):
    if "ERROR" in log:
        print(log)
```

🛡️ Cybersecurity Practice

Analyze suspicious IP addresses.

```python
ip_logs=["192.168.1.10","10.0.0.5","192.168.1.20","172.16.0.3"]
def ip_generator(logs):
    for ip in logs:
        if ip.startswith("192."):
            yield ip
for ip in ip_generator(ip_logs):
    print(ip)
```
