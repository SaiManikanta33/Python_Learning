```python
class Employee:
    company = "Infosys"

    def __init__(self, emp_id, name, salary):
        self.emp_id = emp_id
        self.name = name
        self.salary = salary

    def display(self):
        print(f"Employee ID: {self.emp_id}")
        print(f"Name: {self.name}")
        print(f"Salary: ₹{self.salary}")
        print(f"Company: {Employee.company}")

    def annual_salary(self):
        return self.salary * 12

    def is_performance_bonus_eligible(self):
        return self.salary > 60000

    @classmethod
    def change_company(cls, name):
        cls.company = name


emp = Employee(101, "sai", 100000)
emp.display()
print("Annual Salary:", emp.annual_salary())
print("Eligible for performance bonus:", emp.is_performance_bonus_eligible())

Employee.change_company("TCS")
emp.display()
```

```python
class FirewallRule:
    company="Infosys"
    def __init__(self,rule_name,port):
        self.rule_name = rule_name
        self.port = port
    def display(self):
        print(f"Rule:{self.rule_name}")
        print(f"Port:{self.port}")
        
    @staticmethod
    def is_secure_port(port):
        if port in [22,53,80,443,3389]:
            return port
        else:
            return "UnKnown"
rule = FirewallRule("Allow HTTPS",443)
rule.display()
print(FirewallRule.is_secure_port(22))
```
