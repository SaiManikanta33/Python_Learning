Student Marks Analyzer
Challenge
-Find the highest mark.
-Find the lowest mark.
-Count how many students scored above 90.
-Display grades (A, B, C, Fail) using a function.


```python
marks = [95,82,76,45,68,99,54]
highest_mark= max(marks,key=lambda marks:marks)
low=min(marks,key=lambda marks:marks)
score=list(filter(lambda x:x>90,marks))
grades = list(map(lambda m: "A" if m >= 90 else ("B" if m >= 80 else ("C" if m >= 70 else "Fail")), marks))
print(grades)
print("Highest Marks",highest_mark)
print("Lowest Marks:",low)
print("Students With Above 90 marks:",score)
```
