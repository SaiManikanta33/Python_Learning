Exercise 5

Write a program that:

-Takes two numbers from the user.
-Divides them.
-Handles:
    -Division by zero.
    -Invalid input.

```python
try:
    a = int(input("Enter a value: "))
    b = int(input("Enter b value: "))
    print(a / b)
except ZeroDivisionError:
    print("Cannot divide by 0")


except ValueError:
    print("Invalid input")
```
