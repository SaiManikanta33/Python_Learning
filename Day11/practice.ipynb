Exercise 1

Create a Student class with:

Instance variables:
name
age

Display the information using an instance method.

```python
class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        
    def greet(self):
        print(f"Name:{self.name}")
        print(f"Age: {self.age}")
student1= Student("Sai",19)
student2= Student("Mani",20)

student1.greet()
```

Exercise 2

Create a Company class with:

Class variable: company_name

Create three employees and print the company name for each.

```python
class company:
    company_name1 = "TCS"
    company_name2 = "Infosys"
    
    def __init__(self,name):
        self.name = name
e1 = company("Sai")
e2 = company("Mani")
e3 = company("Manikanta")

print(f"{e1.name} works at {e1.company_name2}")
print(f"{e2.name} works at {e2.company_name1}")
print(f"{e3.name} works at {e3.company_name2}")
```

Exercise 3

Create a class method that changes the company name.

Print the company name before and after the change.

```python
class company:
    company_name = "Infosys"
    
    @classmethod
    def change(cls,name):
        cls.company_name = name
print(company.company_name)
company.change("Amazon")
print(company.company_name)
```

Exercise 4

Create a static method that checks whether a number is even.

```python
class NumberUtils:
    @staticmethod
    def is_even(num):
        return num % 2 == 0
print(NumberUtils.is_even(10))
```

Exercise 5

Create a Rectangle class.

Add a static method:

```python
class rectangle:
    @staticmethod
    def calculate_area(length,width):
        return length * width
print(rectangle.calculate_area(10,20))
```
