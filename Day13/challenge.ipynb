Secure Bank Account
Challenge

Add:

Transaction history (store in a list)
Print the last 5 transactions
Reject deposits of zero or negative amounts

```python
class BankAccount:
    def __init__(self,owner,balance):
        self.owner = owner
        self.__balance = balance
        self.history = []
        
        if balance > 0:
            self.history.append(f"Account history: +₹{balance}")
    def deposit(self,amount):
        if amount <= 0:
            print("The deposit money should greater than 0")
            return
        self.__balance += amount
        self.history.append(f"Deposited: ₹{amount}")
    def withdraw(self,amount):
        if amount <= 0:
            print("The withdraw amount should be greater then 0")
            return
        if amount <= self.__balance:
            self.__balance -= amount
            self.history.append(f"Withdrawl: -₹{amount}")
        else:
            print("Insufficient balance")
    def show_balance(self):
        print(f"Balance: ₹{self.__balance}")
    def show_history(self):
        print(f"-----Recent transactions for {self.owner} -----")
        last_five = self.history[-5:]
        if not last_five:
            print("No transactions ")
            return
        for txn in last_five:
            print(txn)

account = BankAccount("Sai",5000)

account.deposit(1000)
account.withdraw(2000)
account.deposit(10000)
account.withdraw(2000)
account.deposit(1)
account.withdraw(500)
account.show_balance()
account.show_history()
```

Security Alert System
Challenge

Create a ThreatDetector class with:

Private attribute: __threat_level
Getter and setter methods
Reject threat levels outside the range 1–10

```python
from abc import ABC, abstractmethod
class Alert(ABC):
    @abstractmethod
    def notify(self):
        pass
class EmailAlert(Alert):
    def notify(sself):
        print("Email alert sent.")
class SMSAlert(Alert):
    def notify(self):
        print("SMS alert sent.")
class SlackAlert(Alert):
    def notify(self):
        print("💬 Slack alert sent: System monitoring warning.")

class ThreatDetector:
    def __init__(self, threat_level=1):
        self.__threat_level = 1
        self.set_threat_level(threat_level)
    
    def get_threat_level(self):
        return self.__threat_level
    
    def set_threat_level(self, level):
        if 1 <= level <= 10:
            self.__threat_level = level
        else:
            print("Threat level must be between 1 and 10")

detector = ThreatDetector(3)
print(f"Current System Status: Level {detector.get_threat_level()}")
detector.set_threat_level(15)
detector.set_threat_level(8)
print(f"Updated System Status: Level {detector.get_threat_level()}")
if detector.get_threat_level() >= 7:
    print("\n--- CRITICAL CONDITION: Dispatching All Alerts ---")
    alerts = [EmailAlert(), SMSAlert(), SlackAlert()]
    for alert in alerts:
        alert.notify()
```
