Student Record Manager

Requirements:

Store details using a dictionary.
Store completed subjects in a list.
Store favorite port numbers in a set.
Store permanent details (e.g., roll number, admission year) in a tuple.

Example output:

===== Student Profile =====
Name      : Sai Manikanta
Age       : 19
Branch    : Cyber Security

Completed Subjects
- Python
- Linux
- Networking

Favorite Ports
22
80
443

Roll Number : 12345
Admission Year : 2025
===========================

```python
dict={
    "Name":"Sai Manikanta",
    "Age":19,
    "Branch":"Cyber Security"
}
lis=["Python","Linux","Networking"]
set=[22,80,443]
tup=(259854608,2025)
print("========== Student Details ==========")

for name,data in dict.items():
    print(name,"    :",data)
print("\n")
print("Completed Subjects")
for sub in lis:
    print("-",sub)
print("\n")
print("Favourite Ports")
for port in set:
    print(port)
print("\n")
print(f"Roll Number   :{tup[1]}")
print(f"Admission Year   :{tup[1]}")
```

```python
ports = {
    22: "SSH",
    53: "DNS",
    80: "HTTP",
    443: "HTTPS",
    3389: "RDP"
}

for port, service in ports.items():
    print(f"Port {port} -> {service}")
```
