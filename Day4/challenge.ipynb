Simple Login System

Requirements:

Username: admin
Password: python123
Allow 3 attempts.
If credentials are correct, print:
Login Successful

Otherwise:

Invalid Credentials

After 3 failed attempts:

Account Locked

```python
username = input("Enter username:")
passw= input("Enter password")

attempts = 3
while attempts > 0:
    usr=input("Username:")
    pwd = input("Password:")
    if usr == username and pwd == passw:
        print("Login Successfull")
        break
    else:
        attempts -= 1
        print(f"Invalid Credentials. Attempts lefted:{attempts}")
if attempts == 0:
    print("Account Locked")
```

Create a script that checks whether common ports are well-known ports (0–1023).

```python
ports = [22,53,80,443,8080]
for port in ports:
    if port <= 1000:
        print(f"Port {port} is a well-Known Port")
    else:
        print(f"Port {port} is  a registered/Dynamic Port")
```
